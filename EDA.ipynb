### Read in data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv('titanic.csv')
titanic
```

### Explore continuous features

```python
# Drop all categorical features
cat_feat = ['PassengerId', 'Name', 'Ticket', 'Sex', 'Cabin', 'Embarked']
cont_data = titanic.drop(cat_feat, axis=1)
cont_data
```

```python
cont_data.describe()
```

```python
cont_data.groupby('Survived').mean()
```

```python
cont_data.groupby(cont_data['Age'].isnull()).mean()
```

### Plot continuous features

```python
for i in ['Age', 'Fare']:
    died = list(cont_data[cont_data['Survived'] == 0][i].dropna())
    survived = list(cont_data[cont_data['Survived'] == 1][i].dropna())
    xmin = min(min(died), min(survived))
    xmax = max(max(died), max(survived))
    width = (xmax - xmin) / 40
    sns.histplot(died, color='r', kde=False, bins=np.arange(xmin, xmax, width))
    sns.histplot(survived, color='g', kde=False, bins=np.arange(xmin, xmax, width))
    plt.legend(['Did not survive', 'Survived'])
    plt.title('Overlaid histogram for {}'.format(i))
    plt.show()
```

```python
for i, col in enumerate(['Pclass', 'SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=cont_data, kind='point', aspect=2,)
```

```python
cont_data['family_cnt'] = cont_data['SibSp'] + cont_data['Parch']
sns.catplot(x='family_cnt', y='Survived', data=cont_data, kind='point', aspect=2,)
```

### Clean continuous features

```python
# Fill in missing age
titanic.groupby(titanic['Age'].isnull()).mean()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
titanic.isnull().sum()
```
